## Data processing

```python
import os

app = 'sample2'
name = 'eqexec'
folder = os.path.join('/', 'Users', 'snair', 'works', 'dislock-experiments', 'cluster_results', 'new', name, app)

workloads = ['workloadeqeq', 'workloadeqhot', 'workloadeqclust', 'workloadhoteq','workloadhothot','workloadhotclust', 'workloadF', 'workloadG']
gran = [1]
mode = {1:[1,2,3]}
place = {1:[1,2,3]}
runs = [1,2,3,4,5]
replicas = ['paris','tokyo','singapore','capetown','newyork']

wl_values={}
for wl in workloads:
    configs = get_individual_data(folder, wl, gran, mode, place, runs, replicas)
    wl_values[wl] = configs

print(wl_values)
```
